```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
```

```python
start_date =  datetime.datetime(2011, 1, 1)
end_date = datetime.datetime(2021, 12, 31)
```

```python
Tesla = yf.Ticker("TSLA")
TeslaData = (Tesla.history(start = start_date, end = end_date))
```

```python
# To view the first 5 rows in our data set
TeslaData.head()
```

```python
Nvidia = yf.Ticker("NVDA")
NvidiaData = (Nvidia.history(start = start_date, end = end_date))
```

```python
NvidiaData.head()
```

```python
Twitter = yf.Ticker("TWTR")
TwitterData = (Twitter.history(start = start_date, end = end_date))
```

```python
TwitterData.head()
```

For the sake of comparison, I will also pull the NASDAQ 100 index so I can use it to compare with the other 3 stocks.

```python
Nasdaq = yf.Ticker("TWTR")
NasdaqData = (Nasdaq.history(start = start_date, end = end_date))
```

```python
NasdaqData.head()
```

Writing the extracted data into excel for further analysis.

```python
excel = pd.ExcelWriter("StocksDataset.xlsx")
```

```python
TeslaData.to_excel(excel, sheet_name = "TSLA")
```

```python
NvidiaData.to_excel(excel, sheet_name = "NVDA")
```

```python
TwitterData.to_excel(excel, sheet_name = "TWTR")
```

```python
NasdaqData.to_excel(excel, sheet_name = "^NDX")
```

```python
excel.save()
```
